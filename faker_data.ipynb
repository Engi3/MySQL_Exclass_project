```python
import pandas as pd
from faker import Faker
import random
```

สร้างข้อมูลตามจำนวนที่ต้องการ (100)

```python
def mockupdata(row):
    fake = Faker('th_TH')
    data = []
    for i in range(1, row + 1):
        employee = {
            "Employee ID": f"{i:06d}",  # Generate 6-digit employee ID starting from 000001
            "Name": fake.first_name() + ' ' + fake.last_name(),
            "Position": fake.job(),
            "Salary": round(fake.random_number(digits=5), 2),  # Random salary with 5 digits
            "Address": fake.address(),
            "Phone Number": fake.phone_number()
        }
        data.append(employee)

    df = pd.DataFrame(data)
    return df
```

```python
# Generate 100 rows of mockup employee data
df_employees = mockupdata(100)
df_employees.to_csv('employee_data.csv', index=False)
print("Employee data saved to employee_data.csv")
```

```python
# Generate working hours data for each employee for 30 working days
def generate_working_hours_data(df_employees):
    working_days = pd.date_range(start='2023-06-01', end='2023-06-30', freq='B')  # Business days in June 2023 (Mon-Fri)
    working_hours_data = []

    for _, employee in df_employees.iterrows():
        for day in working_days:
            work_entry = {
                "Employee ID": employee["Employee ID"],
                "Name": employee["Name"],
                "Position": employee["Position"],
                "Working Day": day.date(),
                "Start Time": "08:00",
                "End Time": "17:00",
                "OT Time": "20:00" if random.choice([True, False]) else ""  # Randomly assign OT
            }
            working_hours_data.append(work_entry)

    df_working_hours = pd.DataFrame(working_hours_data)
    return df_working_hours
```

```python
df_working_hours = generate_working_hours_data(df_employees)
df_working_hours.to_csv('employee_working_hours.csv', index=False)
print("Working hours data saved to employee_working_hours.csv")

# Display sample data
print(df_working_hours.head())
```
